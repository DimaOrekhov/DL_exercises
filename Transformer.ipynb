## Positional encoding:

```python
from math import cos, sin
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import matplotlib
matplotlib.rcParams.update({'font.size': 20})
```

```python
def encode(position, d):
    vec = []
    for i in range(d//2):
        sine = sin(position/(10000**(2*i/d)))
        cosine = cos(position/(10000**(2*i/d)))
        vec.extend([sine, cosine])
    return vec
```

```python
encode(3, 8)
```

```python
plt.plot([cos(1/(10000**(400/512))*x) for x in np.linspace(0, 10, 100)])
```

```python
arr = []
for pos in range(10):
    arr.append(encode(pos, 256))
arr = np.array(arr)
```

```python
fig, ax = plt.subplots(figsize=[18, 7])
sns.heatmap(arr)
'''
ax.set_yticks([x for x in range(0, 101, 10)])
ax.set_yticklabels([x for x in range(0, 101, 10)])
ax.set_xticks([x for x in range(0, 256, 10)])
ax.set_xticklabels([x for x in range(0, 256, 10)])
'''
ax.set_xlabel('Embedding dimension')
ax.set_ylabel('Position')
```
